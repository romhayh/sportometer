## Creating rectangles using the mouse
click and release when the rectangle is at the right size

````python
import cv2
import numpy as np

isDrawing = False
ix, iy = -1, -1

def draw_rectangle(event, x, y, flags, param):
    global ix, iy, isDrawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        isDrawing = True
        ix, iy = x, y
    elif event == cv2.EVENT_MOUSEMOVE:
        if isDrawing:
            cv2.rectangle(img, (ix, iy), (x, y), (0,255,0), -1)
    elif event == cv2.EVENT_LBUTTONUP:
        isDrawing = False
        cv2.rectangle(img, (ix, iy), (x, y), (0,255,0), -1)


img = np.zeros(shape=(255,255, 3))
cv2.namedWindow(winname='window')
cv2.setMouseCallback('window', draw_rectangle)

while 'displaying image':
    
    
    cv2.imshow('window', img)
    wk2 = cv2.waitKey(1)
    if wk2 & 0xff == ord('q') or wk2 & 0xff == ord('Q') or wk2 & 0xff == 27:
        break
cv2.destroyAllWindows()
````

# Image blending and image pasting

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
dog = cv2.imread(r'./Computer-Vision-with-Python/DATA/dog_backpack.png')
dog = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)
dnc = cv2.imread(r'./Computer-Vision-with-Python/DATA/watermark_no_copy.png')
dnc = cv2.cvtColor(dnc, cv2.COLOR_BGR2RGB)
````

````python
print (fr'dog.shape = {dog.shape}')
print (fr'dnc.shape = {dnc.shape}')
````

````python
dnc = cv2.resize(dnc, (dog.shape[1], dog.shape[0]))
````

````python
print (fr'dog.shape = {dog.shape}')
print (fr'dnc.shape = {dnc.shape}')
````

````python
#blend both images
blended = cv2.addWeighted(dog, 1, dnc, 1, -256)
plt.imshow(blended)
````

## blending images that have different sizes:
lets start by pasting an image to a bigger one


````python
# change dnc`s size 
dog_copy = np.copy(dog)
sdnc = cv2.resize(dnc, (400, 400))
````

````python
# add offset to be able to slice in a more convenient way
xoff = 200 
xend = xoff + sdnc.shape[1]
yoff = 400
yend = yoff + sdnc.shape[0]
````

````python
# paste the smaller image onto the bigger one, in the region of intrest
roi = dog_copy[yoff:yend, xoff:xend]
plt.imshow(roi)
dog_copy[yoff:yend, xoff:xend] = sdnc
````

````python
plt.imshow(dog_copy)
````

now lets blend it all together

````python
dog_copy = np.copy(dog)
roi = dog_copy[yoff:yend, xoff:xend]
plt.imshow(roi) 
````

to be able to blend two images, we need a mask  
in this example, the mask will be set to the do not copy Image

````python
mask = np.copy(sdnc)
mask = cv2.cvtColor(mask, cv2.COLOR_RGB2GRAY)
mask = cv2.bitwise_not(mask)
plt.imshow(mask, cmap='gray')
````

### making the mask available in rgb
#### V1

````python
white_bg = np.ones(shape=sdnc.shape) * 255
````

````python
mask = cv2.bitwise_or(white_bg,white_bg, mask=mask)
````

````python
mask.shape
````

````python
plt.imshow(mask)
````

#### V2

````python
mask = np.copy(sdnc)
mask = cv2.cvtColor(mask, cv2.COLOR_RGB2GRAY)
mask = cv2.bitwise_not(mask)
mask_inv = mask
plt.imshow(mask, cmap='gray')
````

````python
temp = np.zeros(sdnc.shape)
````

````python
for i in range(3):
    temp[:,:,i] = mask
mask = temp
````

````python
plt.imshow(mask)
````

add the mask to the small image

````python
sdnc_mask = np.copy(sdnc)
sdnc_mask = cv2.bitwise_or(sdnc_mask,sdnc_mask, mask=mask_inv)
plt.imshow(sdnc_mask)
````

add the final mask onto the picture

````python
dog_copy = np.copy(dog)
dog_copy[yoff:yend, xoff:xend] = cv2.bitwise_or(dog_copy[yoff:yend, xoff:xend], sdnc_mask)
plt.imshow(dog_copy)
````

### different types of thresholds

#### binary

````python
ret, img_t = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
````

````python
plt.imshow(img_t, 'gray')
````

#### inverted binary

````python
ret, img_t = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(img_t, 'gray')
````

#### truncate

````python
ret, img_t = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
plt.imshow(img_t, 'gray')
````

#### to zero

````python
ret, img_t = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
plt.imshow(img_t, 'gray')
````

#### to zero inverted

````python
ret, img_t = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(img_t, 'gray')
````

# show_pic
### show the picture in a more convenient size in the notebook

````python
def show_pic(img, cmap=None, (w, h) = (20, 20)):
    fig = plt.figure(figsize=(w,h))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap=cmap)
````

````python
img = cv2.imread(r'./Computer-Vision-with-Python/DATA/crossword.jpg',0)
````

````python
show_pic(img, 'gray')
````

#### lets scan the paper

````python
ret, img_t = cv2.threshold(img,200, 255, cv2.THRESH_BINARY)
show_pic(img_t, 'gray')
````

````python
#using adaptive threshold
img_t = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 9)
show_pic(img_t, 'gray')
````
